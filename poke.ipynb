```python
import pandas as pd
import random
from datetime import datetime, timedelta

# 시작 날짜와 끝 날짜 지정
start_date = datetime(2024, 10, 1)
end_date = datetime(2024, 10, 31)

# 지정된 기간 내에서 랜덤 날짜 생성
def random_date(start, end):
    return start + timedelta(seconds=random.randint(0, int((end - start).total_seconds())))

# 랜덤 row를 생성하는 함수
def generate_random_row(start, end):
    rand_datetime = random_date(start, end)  # 랜덤 datetime 생성
    rand_date = rand_datetime.date()         # date 추출
    rand_timestamp = int(rand_datetime.timestamp())  # unix timestamp로 변환
    return [rand_date, rand_datetime, rand_timestamp]

# 지정된 수만큼 랜덤 row를 생성하여 DataFrame 반환
def generate_random_dataframe(n, start, end):
    rows = [generate_random_row(start, end) for _ in range(n)]
    df = pd.DataFrame(rows, columns=['date', 'datetime', 'unix timestamp'])
    return df

# 예시: 10개의 랜덤 row 생성
df = generate_random_dataframe(10, start_date, end_date)

# DataFrame 출력
print(df)
```
